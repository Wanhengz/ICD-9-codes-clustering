```python
def find_all_types(contents):
    types = []
    lines = contents.split('\n')
    concept_lines = [line for line in lines if line.startswith('  ',0)]
    for line in concept_lines:    
        start = line.index('[')+1
        end = line.index(']')
        t = line[start:end]
        types.append(t)
    unique_types = list(set(types))
    return(types,unique_types)
```

```python
def create_semantic_dictionary(contents):
    from collections import defaultdict
    dic = defaultdict(list)
    #types = find_all_unique_types(contents)
    #dic = dict(zip(types, [None]*len(types)))
    lines = contents.split('\n')
    concept_lines = [line for line in lines if line.startswith('  ',0)]
    for line in concept_lines:
        value = line.split('[')[0].split('(')[0][9:-1]
        key = line.split('[')[1][:-1]
        if value not in list(dic[key]): 
            dic[key].append(value)
    dic = dict(dic)
    return dic
```

```python
def create_semantic_table(contents,icd9,des):
    #create a empty table with row names and column names
    index = [index for index in range(len(contents)) if contents.startswith('Processing',index)]
    ctt = []
    col_list = []
    e = create_semantic_dictionary(contents)
    for key,values in e.items():
        for value in values:
            y = (key,value)
            col_list.append(y)
    row = des
    col= pd.MultiIndex.from_tuples(col_list)
    table = pd.DataFrame(None, row,col)
    # filling the table with indicator
    for i in range(len(index)-1):
        start = index[i]
        end = index[i+1]
        a = contents[start:end]
        ctt.append(a)
    ctt.append(contents[index[-1]:])
    # filling the table with indicator
    for i in range(len(ctt)):
        item = ctt[i]
        lines = item.split('\n')
        concept_lines = [line for line in lines if line.startswith('  ',0)]
        for line in concept_lines:
            category = line.split('[')[1][:-1]
            concept = line.split('[')[0].split('(')[0][9:-1]
            des = des[i]
            table.loc[des,(category,concept)] = 1
    return(table)
```
